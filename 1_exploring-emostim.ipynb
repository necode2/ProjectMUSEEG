```python
#libraries
import pandas as pd
import numpy as np
```

```python
stim_vids = pd.read_csv("/kaggle/input/filmclipsdetailscsv/FilmClipsDetails(in).csv") 
```

```python
print(stim_vids.columns.tolist())

"""
'Interest'--> 'positive'
'Fear'--> 'negative'
'Anxious' --> 'negative'
'Moved'--> 'positive'
'Anger' --> 'negative'
'Ashamed' --> 'negative'
'Warm-hearted'--> 'positive'
'Joy'--> 'positive'
'Sad' --> 'negative'
'Satisfied' --> 'positive'
'Surprise' --> 'delete'
'Love' --> 'positive'
'Guilt' --> 'negative'
'Disgust' --> 'negative'
'Disdainful' --> 'negative'
'Calm'--> 'positive'
"""
```

```python
highest_emotion = 0
lowest_emotion = 0
column_name = 'column name'
emotion_list = []
valid_columns = {'Interest':'positive',
                  'Fear': 'negative',
                'Anxious' : 'negative',
                         'Moved': 'positive',
                         'Anger' : 'negative',
                         'Ashamed' : 'negative',
                         'Warm-hearted': 'positive',
                         'Joy': 'positive',
                         'Sad' : 'negative',
                         'Satisfied' : 'positive',
                         'Love' : 'positive',
                         'Guilt' : 'negative',
                         'Disgust' : 'negative',
                         'Disdainful' : 'negative',
                         'Calm': 'positive'}
binary_list = []
n_num = 0
p_num = 0

for index, row in stim_vids.iterrows():
    print(f"row index:{index}")
    for column, value in row.items():
        if column in valid_columns:
            #print(value)
            #if value > highest_emotion:
            if valid_columns[column] == "positive":
                highest_emotion+=value
            if valid_columns[column] == "negative":
                lowest_emotion+=value
    final_posRating = highest_emotion / 35 # 7 possible pos * 5 possible rating
    final_negRating = lowest_emotion / 40 # 8 possible neg * 5 possible rating
    
    print(f"positive emotion: {final_posRating}, negative emotion: {final_negRating}")
    if final_posRating > final_negRating:
        binary_list.append("positive")
        p_num+=1
    elif final_posRating < final_negRating:
        binary_list.append("negative")
        n_num+=1
    
    highest_emotion = 0
    lowest_emotion = 0

print("-------------------------")
print(f"positive num: {p_num}")
print(f"negative num: {n_num}")
print(binary_list)
print(f"binary list size: {len(binary_list)}")
```

```python
relevant_info = stim_vids[['clip_name','video_url', 'used_in_study', 'clip_duration']].copy()
relevant_info.head()
```

```python
# adding the associated pos/neg rating 
relevant_info['rating'] = binary_list
relevant_info.head()
```

```python
# remove videos not used in study 
relevant_info = relevant_info[relevant_info['used_in_study'] != 'No']
relevant_info.head()

relevant_info = relevant_info.drop(columns=['used_in_study'])

n_num = 0
p_num = 0
for rating in relevant_info['rating']:
    if rating == 'negative':
        n_num += 1
    if rating == 'positive':
        p_num += 1

print(f"positive num: {p_num}")
print(f"negative num: {n_num}")
print(f"total num movies: {relevant_info['rating'].size}")
print("------------------------------------")
relevant_info.head()
```

```python
# save updated csv
relevant_info.to_csv('relevant_vidData.csv', index=False)
```
